```python
import os
import numpy as np
import pandas as pd
import pickle as pkl
import random
```

# 오리지널 데이터

```python
data_df = pd.read_csv("../data/train/train.tsv", sep="\t", header=None)
data_df
```

```python
with open("../data/label_type.pkl", "rb") as lt:
    label_encoder = pkl.load(lt)
```

```python
label_counts = {k: data_df[8].value_counts()[k] for k in label_encoder.keys()}
label_counts
```

# 추가 데이터

```python
extra_data = pd.read_csv("../data/train/extras.tsv", sep="\t", header=None)
# extra_data_1 = pd.read_csv("../data/train/19_bornIn_city.tsv", sep="\t", header=None)
# extra_data_2 = pd.read_csv("../data/train/26_bornIn_country.tsv", sep="\t", header=None)
# extra_data_3 = pd.read_csv("../data/train/37_dienIn_city.tsv", sep="\t", header=None)
# extra_data_4 = pd.read_csv("../data/train/40_dienIn_country.tsv", sep="\t", header=None)
# extra_data = pd.concat([extra_data_0, extra_data_1, extra_data_2, extra_data_3, extra_data_4])
extra_data
```

```python
extra_counts = extra_data[8].value_counts()
extra_counts
```

# 데이터 추출하기

```python
max_pick_num = int(np.mean(list(label_counts.values())) * 1.2)
max_pick_num
```

```python
tgt_data = pd.DataFrame(columns=data_df.columns)
for k, v in extra_counts.items():
    if max_pick_num - label_counts[k] <= 0:
        continue
    pick_num = min(max_pick_num - label_counts[k], extra_counts[k])
    sample = np.random.choice(extra_data.loc[extra_data[8] == k, :].index, size=pick_num)
    pick_data = extra_data.iloc[sample, :]
    tgt_data = pd.concat([tgt_data, pick_data])
tgt_data
```

```python
tgt_data[8].value_counts()
```

```python
augmented_data = pd.concat([data_df, tgt_data])
augmented_data.to_csv("../data/train/augmented.tsv", sep="\t", header=None, index=False)
```
